# **Components of a Figure in Matplotlib**

A Figure in Matplotlib is the complete window or canvas that contains one or more plots. Understanding its components helps in creating and customizing visualizations effectively.

Main Components of a Figure

1. Figure
* The top-level container for all plot elements.
* Represents the entire drawing area.

2. Axes
* The actual plotting area where data is drawn.
* A figure can contain multiple axes (subplots).

3. Axis
* Defines the coordinate system.
* Consists of:
    * X-Axis (horizontal)
    * Y-Axis (vertical)

4. Title
* Describes the plot.

5. Labels
* Explain what each axis represents.

6. Ticks
* Marks and values displayed along the axes.

7. Legend
* Identifies different data series.

8. Grid
* Reference lines that make data easier to read.

# **Matplotlib: Pyplot API vs Object-Oriented API**

Matplotlib provides two main ways to create plots:

1. Pyplot API (State-Based Interface)
2. Object-Oriented (OO) API

## **Pyplot API**

The matplotlib.pyplot module provides MATLAB-like plotting functions.

Features:
* Easy to learn and use
* Best for simple and quick plots
* Uses the current figure and axes automatically

````python
x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
````

**How It Works**

Pyplot -> Current Figure -> Current Axes -> Plot

## **Object-Oriented (OO) API**

In the Object-Oriented approach, you explicitly create and manage Figure and Axes objects.

Features:
* More control over plots
* Recommended for complex visualizations
* Easier to manage multiple subplots

````python
fig, ax = plt.subplots()

x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

ax.plot(x, y)
ax.set_title("Line Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")

plt.show()
````

**How It Works**

Figure Object -> Axes Object -> Plot Methods

# **Matplotlib Plotting**

## **Plotting x and y points**

The **plot() function** is used to draw points (markers) in a diagram.

By default, the plot() function draws a line from point to point.

The function takes parameters for specifying points in the diagram.

**Parameter 1** is an array containing the points on the **x-axis**.

**Parameter 2** is an array containing the points on the **y-axis**.

````python
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints)

plt.show()
````

The **x-axis** is the horizontal axis.

The **y-axis** is the vertical axis.

## **Plotting Without Line**

To plot only the markers, you can use shortcut string notation parameter 'o', which means 'rings'.



````python
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, "o")

plt.show()
````

## **Multiple Points**

You can plot as many points as you like, just make sure you have the same number of points in both axis.

````python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)

plt.show()
````

## **Default X-Points**

If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 etc., depending on the length of the y-poin

````python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)

plt.show()
````

# **Matplotlib Labels and Title**

## **Create Labels for a Plot**

With Pyplot, you can use the **xlabel()** and **ylabel**() functions to set a label for the x- and y-axis.



````python
avg_pulse = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
cal_burnage = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(avg_pulse, cal_burnage)

plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
````

## **Create a Title for a Plot**

With Pyplot, you can use the **title()** function to set a title for the plot.

````python
avg_pulse = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
cal_burnage = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(avg_pulse, cal_burnage)
plt.show()
````

## **Set Font Properties for Title and Labels**

You can use the fontdict parameter in xlabel(), ylabel(), and title() to set font properties for the title and labels.

````python
avg_pulse = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
cal_burnage = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(avg_pulse, cal_burnage)
plt.show()
````

# **Matplotlib Subplot**

## **Display Multiple Plots**

**subplot()** is used to display multiple plots in the same figure.

**Syntax:** plt.subplot(rows, columns, index)
* rows → Number of rows in the figure
* columns → Number of columns in the figure
* index → Position of the current plot (starts from 1)

````python
x = np.array([1, 2, 3, 4])
y1 = np.array([1, 4, 9, 16])
y2 = np.array([1, 8, 27, 64])

plt.subplot(1, 2, 1)  # 1 row, 2 columns, plot 1
plt.plot(x, y1)
plt.title("Square")

plt.subplot(1, 2, 2)  # 1 row, 2 columns, plot 2
plt.plot(x, y2)
plt.title("Cube")

plt.show()
````

````python
x = np.array([1, 2, 3, 4])

plt.subplot(2, 2, 1)
plt.plot(x, x)
plt.title("Linear")

plt.subplot(2, 2, 2)
plt.plot(x, x**2)
plt.title("Square")

plt.subplot(2, 2, 3)
plt.plot(x, x**3)
plt.title("Cube")

plt.subplot(2, 2, 4)
plt.plot(x, x**4)
plt.title("Power 4")

plt.tight_layout()
plt.show()
````

## **Super Title**

You can add a title to the entire figure with the suptitle() function:

````python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle("MY SHOP")
plt.show()
````

# **Matplotlib Box Plot**

A Box Plot (or Box-and-Whisker Plot) is used to summarize the distribution of numerical data and detect outliers.

It shows:
* Minimum value
* First Quartile (Q1)
* Median (Q2)
* Third Quartile (Q3)
* Maximum value
* Outliers

````python
data = [40, 45, 50, 55, 60, 65, 70, 75, 80, 100]

plt.boxplot(data)

plt.title("Box Plot")

plt.show()
````

**Parts of a Box Plot**

Consider the sorted data:
40 45 50 55 60 65 70 75 80 100

1. Minimum Value:
- Smallest value: 40

2. First Quartile (Q1)
- 25% of data lies below Q1: Q1 = 50

3. Median (Q2)
- Middle value : Median = 62.5

4. Third Quartile (Q3)
- 75% of data lies below Q3 : Q3 = 75

5. Maximum Value
- Largest value: 100

## **Structure of a Box Plot**

```
Min      Q1      Median      Q3      Max
 |--------|====|====|--------|
          Box
```

```
      ┌───────────┐
------│     |     │------
      └───────────┘
```
* Left line = Minimum
* Left edge of box = Q1
* Middle line = Median
* Right edge of box = Q3
* Right line = Maximum







````python
data = [10, 12, 13, 15, 16, 18, 20, 22, 100]

plt.boxplot(data)

plt.show()
````

## **Horizontal Box Plot**

````python
data = [10, 12, 13, 15, 16, 18, 20, 22, 100]

plt.boxplot(data, vert=False)

plt.show()
````

## **Multiple Box Plots**

Compare several datasets:

````python
math = [60, 70, 75, 80, 85]
science = [50, 60, 65, 70, 75]
english = [55, 65, 70, 75, 90]

plt.boxplot([math, science, english])

plt.xticks([1, 2, 3],

           ["Math", "Science", "English"])

plt.show()
````

# **Matplotlib Pie Charts**

## **Creating Pie Charts**

With Pyplot, you can use the pie() function to draw pie charts:



````python
y = np.array([35, 25, 25, 15])

plt.pie(y)

plt.show()
````

**Note:** The size of each wedge is determined by comparing the value with all the other values, by using this formula:

**The value divided by the sum of all values: x/sum(x)**

## **Labels**

Add labels to the pie chart with the labels parameter.

The **labels** parameter must be an array with one label for each wedge:

````python
y = np.array([35, 25, 25, 15])

mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)

plt.show()
````

## **Start Angle**
As mentioned the default start angle is at the x-axis, but you can change the start angle by specifying a startangle parameter.

The startangle parameter is defined with an angle in degrees, default angle is 0:

````python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
````

## **Explode**

Maybe you want one of the wedges to stand out? The explode parameter allows you to do that.

The explode parameter, if specified, and not None, must be an array with one value for each wedge.

Each value represents how far from the center each wedge is displayed:



````python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
````

````python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, explode = [.25, .5, 0, 0])
plt.show()
````

## **Shadow**

Add a shadow to the pie chart by setting the shadows parameter to True:

````python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, explode = [.25, 0, 0, 0], shadow = True)
plt.show()
````

## **Legend**

To add a list of explanation for each wedge, use the **legend()** function:



````python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show()
````

## **Legend With Header**

To add a header to the legend, add the title parameter to the legend function.

````python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show()
````

# **Count Plot in Matplotlib**

A Count Plot shows the number of occurrences (frequency) of each category.

Matplotlib does not have a built-in countplot() function.
The **countplot() function** belongs to the  Seaborn documentation⁠￼ library.

In Matplotlib, we create a count plot using **bar()**.

````python
from collections import Counter

departments = ["CSE", "ECE", "CSE", "ME", "CSE", "ECE", "CSE", "CE", "ME", "CSE", "ECE", "CE", "CSE"]

count = Counter(departments)

plt.bar(count.keys(), count.values())

plt.title("Student Count by Department")
plt.xlabel("Department")
plt.ylabel("Number of Students")

plt.show()
````

# **Matplotlib Scatter**

## **Creating Scatter Plots**

With Pyplot, you can use the scatter() function to draw a scatter plot.

The **scatter()** function plots one dot for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis:



````python
car_age = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
car_speed = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(car_age, car_speed)

plt.show()
````

## **Compare Plots**

````python
#day one, the age and speed of 13 cars:
car_age = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
car_speed = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(car_age, car_speed)

#day two, the age and speed of 15 cars:
car_age = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
car_speed = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])

plt.scatter(car_age, car_speed)

plt.show()
````

**Note:** The two plots are plotted with two different colors, by default blue and orange, you will learn how to change colors later in this chapter.

## **Colors**

You can set your own color for each scatter plot with the color or the c argument:

````python
#day one, the age and speed of 13 cars:
car_age = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
car_speed = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(car_age, car_speed, color = "hotpink")

#day two, the age and speed of 15 cars:
car_age = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
car_speed = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(car_age, car_speed, color = "#ff0000")

plt.show()
````

## **Size**

You can change the size of the dots with the s argument.

Just like colors, make sure the array for sizes has the same length as the arrays for the x- and y-axis:

````python
car_age = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
car_speed = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(car_age, car_speed, s = sizes)

plt.show()
````

Just like colors, make sure the array for sizes has the same length as the arrays for the x- and y-axis:

````python
car_age = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
car_speed = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(car_age, car_speed, s = sizes, alpha = 0.5)

plt.show()
````

# **3D Plot in Matplotlib**

A 3D Plot is used when data has three variables:
* X-axis → Variable 1
* Y-axis → Variable 2
* Z-axis → Variable 3

## **Import 3D Module**


````python
from mpl_toolkits.mplot3d import Axes3D
````

## **3D Scatter Plot**

````python
age = [25, 30, 35, 40, 45]
salary = [30000, 40000, 50000, 60000, 70000]
experience = [1, 3, 5, 8, 10]

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(age, salary, experience)

ax.set_xlabel("Age")
ax.set_ylabel("Salary")
ax.set_zlabel("Experience")

ax.set_title("3D Scatter Plot")

plt.show()
````

## **3D Line Plot**

````python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
z = [1, 4, 9, 16, 25]

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

ax.plot(x, y, z)

ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")

ax.set_title("3D Line Plot")

plt.show()
````
